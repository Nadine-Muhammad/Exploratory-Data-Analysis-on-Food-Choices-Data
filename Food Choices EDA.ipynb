# the cleaning code

```python
data = data_original.copy()
```

```python
#replacing 1 and 2 with their alternatives to make dataset clear
data.coffee = data.coffee.replace(1,'creamy frapuccino')
data.coffee = data.coffee.replace(2,'espresso')
data.Gender = data.Gender.replace(1,'Female')
data.Gender = data.Gender.replace(2,'Male')
data.breakfast = data.breakfast.replace(1,'Cereal')
data.breakfast = data.breakfast.replace(2,'Donut')
data.drink = data.drink.replace(1,'Orange Juice')
data.drink = data.drink.replace(2,'Soda')
data.vitamins = data.vitamins.replace(1,'Yes')
data.vitamins = data.vitamins.replace(2,'No')
```

```python
data
```

```python
#the column comfort_food_reasons_coded is repeated
data = data.drop('comfort_food_reasons_coded', axis=1)
```

```python
#accessing the meaningless non numerical data in GPA and removing them
data.GPA = data.GPA.replace('Unknown', None)
data.GPA = data.GPA.replace('Personal ', None)
#removing noise from data
data.GPA = data.GPA.replace('3.79 bitch',3.79)
```

# classification

```python
import numpy as np
```

```python
data['GPA'] =data['GPA'].astype('float32')
data['weight'] = data['weight'].astype('float32')
data['self_perception_weight'] = data['self_perception_weight'].astype('float32')
data['eating_changes_coded'] = data['eating_changes_coded'].astype('float32')
data['eating_out'] = data['eating_out'].astype('float32')
data.Gender = data.Gender.replace('Female',1)
data.Gender = data.Gender.replace('Male',2)
```

```python
conditions = [
    (data['weight'] >= 157.886957),
    ((data['weight'] < 177.500000) & (data['weight'] >= 155.000000)),
    (data['weight'] < 135.000000)]

# create a list of the values we want to assign for each condition
values = ['Class_1', 'Class_2','Class_3']

# create a new column and use np.select to assign values to it using our lists as arguments
data['weight_Category'] = np.select(conditions, values)
```

```python
from sklearn import tree
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split function
from sklearn.model_selection import train_test_split 
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

X = data[['Gender','self_perception_weight','eating_changes_coded','eating_out','GPA']]
Y = data.weight_Category

# 70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) 
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion='gini',splitter='random')

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred) * 100)
```

```python
#Visualizing the results
tree.plot_tree(clf);
```
